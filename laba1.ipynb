# 2.3 Цикл

Цикл `while` («пока») позволяет выполнить одну и ту же последовательность действий, пока проверяемое условие истинно. Условие записывается до тела цикла и проверяется до выполнения тела цикла. Как правило, цикл `while` используется, когда невозможно определить точное значение количества проходов исполнения цикла.

Синтаксис цикла `while`:

```
while <условие>:
    <блок инструкций>
```

Цикл `for`. Синтаксис:

```
for <переменная итерирования> in range(<начало полуинтервала (включительно)>, 
                                       <конец полуинтервала (не включительно)>, 
                                       <шаг>):
    <блок инструкций>
```

Два **сниппета** (два куска кода) ниже эквивалентны:

```
for i in range(10):
    <блок инструкций>
```

и

```
i = 0
while i < 10:
    <блок инструкций>
    i += 1
```

Функция `range()` работает в зависимости от количества предоставленных аргументов:

1. Если аргумент один, то `i` (переменная итерирования) проходит с шагом 1 по полуинтервалу `[0; <аргумент>)`
2. Если аргумента два, то `i` проходит с шагом 1 по полуинтервалу `[<аргумент 1>; <аргумент 2>)`
3. Если аргумента три, то `i` проходит с шагом `<аргумент 3>`:
    - если `<аргумент 3>` больше 0, то — по полуинтервалу `[<аргумент 1>; <аргумент 2>)`
    - если `<аргумент 3>` меньше 0, то — по полуинтервалу `(<аргумент 2>; <аргумент 1>]`, начиная с конца.
    
Все аргументы `range()` - целые числа.

# Пример 1. "Метание мяча"

Рассмотрим пример, иллюстрирующий математическую модель, описывающую траекторию полета мяча в воздухе. По второму закону Ньютона, предполагая отсутствие сопротивления воздуха, мы можем вывести зависимость вертикального положения мяча от времени $t$:


$$y(t)=v_0t−\frac{1}{2}gt^2$$


где $v_0$ — начальная скорость, $g$ — ускорение свободного падения, значение которого положим равным 9.80665 $м/c^2$.

(Обратите внимание, какие красивые формулы! Текстовые ячейки Jupyter поддерживают набор формул в синтаксисе вёрстки LaTeX, наиболее популярного набора макрорасширений (или макропакета) системы компьютерной вёрстки TeX.)

# Сценарий

Рассмотрим сценарий на Python для вычисления по этой простой формуле

````python
# Программа для вычисления положения мяча при вертикальном движении

v0 = 5       # Начальная скорость
g = 9.80665  # Ускорение свободного падения
t = 0.6      # Время

y = v0 * t - .5 * g * t ** 2 # '.5' - один из способов записи '0.5'

print(y)
````

# Разбор сценария, или "Как он это сделал?"

Сценарий на языке Python — это текст (или текстовый файл, который можно запустить в редакторе Spyder или в консоли Python), содержащий некоторые инструкции. Мы можем читать сценарий и понимать, что программа способна делать, но сам сценарий не выполняет никаких действий на компьютере, пока интерпретатор Python не прочитает текст сценария и не преобразует его в некоторые действия.

Когда сценарий запущен в интерпретаторе Python, он разбирается и выполняется построчно. То, где вы это читаете, не является частью сценария.

Первая строчка:

````python
# Программа для вычисления положения мяча при вертикальном движении
````

является комментарием: как только встречается символ `#`, интерпретатор Python воспринимает оставшуюся часть строки как комментарий, пропускает её и переходит к следующей строке.

Следующая строка, интерпретируемая Pyhton:

````python
v0 = 5       # Начальная скорость
````

В Python выражения вида `v0 = 5` известны как операции присваивания (с оператором "="). Значение правой части (в нашем случае — целое число 5) становится объектом, а имя переменной слева — именованной ссылкой на этот объект. Всякий раз, когда мы запишем `v0`, Python заменит ее целым значением 5. Выражение `v1 = v0` создает новое имя для того же целого объекта со значением 5, а не копирует объект. 

Следующие две строки:

````python
g = 9.80665  # Ускорение свободного падения
t = 0.6      # Время
````

имеют тот же смысл, что и предыдушая. Таким образом, прочитав эти строки, интерпретатор Python знает три переменных `v0`, `g`, `t` и их значения. Эти переменные используются интерпретатором в следующей строке, реализующей некоторую формулу:

````python
y = v0 * t - .5 * g * t ** 2 # '.5' - один из способов записи '0.5'
````

В этой строке Python интерпретирует `*` как оператор умножения, `-` — вычитания, `**` — возведения в степень (вместо `^`). Естественно, `+` и `\` интерпретируются как операторы сложения и деления соответственно. В результате вычисляется значение по формуле в правой части выражения, которое присваивается переменной с именем `y`. 

Последняя строка сценария:

````python
print(y)
````

выводит на печать значение переменной `y`. Таким образом, при запуске сценария будет выведено число `1.234803`. В тексте сценария имеются также и пустые строки, которые пропускаются интерпретатором Python. Они добавлены для лучшей читабельности кода.


# Упражнение 4. Проверка корректности исходных данных

Заметим, что в рассматриваемой задаче ось $y$ выбрана так, что мячик в начальный момент времени ($t = 0$) находится в точке с координатой $y = 0$. Чтобы посмотреть, сколько времени займет у мячика вернуться в эту же координату ($y = 0$), надо рассмотреть следующее выражение:

$$v_0t−\frac{1}{2}gt^2 = t \left( v_0t−\frac{1}{2}gt \right) = 0$$

Последнее возможно, только если: $$t = 0$$ или: $$t = \frac{2v_0}{g}.$$

Таким образом, возможное время полета мяча ограничено интервалом:

$$0 \leqslant t \leqslant \frac{2v_0}{g}.$$

**Задание:** добавить в программу проверку корректности вводимых пользователем переменных, исходя из условий: 

- начальная скорость и ускорение свободного падения не отрицальны
- диапазон возможного времени определяется полученным выше условием.

````python
g  = 9.80665  # Ускорение свободного падения (м/с^2)

v0 = int(input('Введите начальную скорость (в метрах в секунду): '))
while not isinstance(v0, (int, float)) or v0 < 0:
    v0 = int(input('Введите начальную скорость (в метрах в секунду): '))

t  = int(input('Введите время (в секундах): '))
while not isinstance(t, (int, float)) or not (0 <= t <= (2*v0/g)):
    t  = int(input('Введите время (в секундах): '))
    
y  = v0 * t - .5 * g * t ** 2 # '.5' - один из способов записи '0.5'

print('\nИсходные данные:')
print('%-30s %.5e м/с' % ('Начальная скорость', v0))
print('%-30s %.5e м/(с^2)' % ('Ускорение свободного падения', g))
print('%-30s %.5e с' % ('Время', t))
print('\nВыходные данные:')
print('%-30s %.5e м' % ('Текущая высота положения мяча', y))
````

После введения каждого нового числа пересчитывать и выводить результат в формате: "Среднее значение этих (число чисел) чисел равно" + результат работы с точностью до 5 знаков после запятой (т.е. до 1/100000).
````python
x = 1 + 2 + 3 + 4 + 5
num = 5
sr_x = x / num
print('Среднее значение этих 5 чисел равно %.5f' % sr_x)

i = int(input('Введите новое число: '))
if i != 0:
    while i != 0:
        num += 1
        x += i
        sr_x = x / num
        print('Среднее значение этих %i чисел равно %.5f' % (num, sr_x))
        i = int(input('Введите новое число: '))
````

Рассмотрим пример иллюстрирующий математическую модель описывающий траекторию кошачьего прыжка. Из второго закона Ньютона, предполагая отсутствие сопротивления воздуха, мы можем вывести зависимость вертикального (y) и горизонтального (x) положения котика от времени $t$:

$$x(t)=v_0cos(\alpha)t$$
$$y(t)=v_0sin(\alpha)t−\frac{1}{2}gt^2$$

где $v_0$ — начальная скорость, $g$ — ускорение свободного падения, значение которого положим равным 9.80665 $м/c^2$, $tg(\alpha) = y/x$.

Напишите программу которая сможет помочь котику понять следующие вещи:


1) как далеко и высоко он сможет прыгнуть при начальной скорости 5 м/с и направлении пражка $\alpha = 40^{\circ}$;

2) каким должен быть начальный угол для максимальной дальности прыжка;

3) каким должен быть начальный угол для максимальной высоты прыжка;

4) каким должны быть угол и скорость для попадания в точку с координатами (2,3) м через 1 с;

5) нарисуйте траекторию прыжка с помощью библиотеки matplotlib.


````python
from math import sin, acos, cos
import pylab

pi = 3.14159
g = 9.80665

def x_t(v0, a, t):
    return v0 * cos(a) * t

def y_t(v0, a, t): 
    return (v0 * sin(a) * t) - (0.5 * g * (t ** 2))
````

````python
v0 = 0
v0 = input('Введите начальную скорость (по умолчанию - 5 м/с):')
if v0 == '':
    v0 = 5
v0 = float(v0)

alpha = 0
alpha = input('Введите угол прыжка (по умолчанию - 40 градусов к горизонту):')
if alpha == '':
    alpha = 40
alpha = float(alpha)
    
alpha = alpha * pi / 180
    
h_max = (v0 * (sin(alpha) ** 2)) / (2*g)
l_max = ((v0 ** 2) * sin(alpha * 2)) / g

print('Максимальная высота прыжка равна %f м, а максимальная длина - %f м' % (h_max, l_max))
````

````python
l = []
for a in range(91):
    l.append(v0 ** 2 * sin(2 * a * pi / 180) / g)
print('Начальный угол для максимальной дальности прыжка равен:' , "%d" % l.index(max(l)) + '°')
````

````python
l = []
for a in range(91):
    l.append(v0 ** 2 * sin(a * pi / 180) / g)
print('Максимальная высота полета тела достигается при угле вылета, равном:', "%d" % l.index(max(l)) + '°')
````

````python
x = 0
x = input('Введите x (по умолчанию - 2 м):')
if x == '':
    x = 2
x = float(x)
    
y = 0
y = input('Введите y (по умолчанию - 3 м):')
if y == '':
    y = 3
y = float(y)
    
t = 0
t = input('Введите t (по умолчанию - 1 c):')
if t == '':
    t = 1
t = float(t)
    
v0 = ( (y + 0.5*g*(t ** 2)) ** 2 + x ** 2 / (t ** 2) ) ** 0.5
print('Скорость равна %.5f м/с.' % v0)

alpha = acos(x / (v0 * t))
print('Угол равен %.5f радиан, что в градусах равняется %.5f.' % (alpha, alpha * 180 / pi))
````

````python
t_sh = 0
t_sh = input('Введите шаг точности построения (по умолчанию - 0.1 c):')
if t_sh == '':
    t_sh = 0.1
t_sh = float(t_sh)

print('По последним известиям график представлен ниже:')

xlist = []
ylist = []

t0 = 0

while t0 < t:
    xlist.append(x_t(v0, alpha, t0))
    ylist.append(y_t(v0, alpha, t0))
    t0 += t_sh
    
pylab.plot(xlist, ylist)
pylab.show()
````

# Задание. "Число и цифра"

Запросите у пользователя целое число и сложите цифры этого числа. Повторяйте последнее действие столько раз, сколько сможете, и выводите каждый раз полученную сумму.

````python
x = int(input('Введите целое число: '))
s = 0

while x > 9:
    while (x // 10) > 0:
        s += x % 10
        x = x // 10
    x = s + x
    s = 0
    print('Текущая сумма равна %i' % x)
````

# Задание. "Максимум и минимум"

Запрашивайте у пользователя числа, определяйте максимальное и минимальной из всех полученных чисел и сообщайте их ему.

````python
mx, mi = 0, 0

num = int(input('Введите количество чисел: '))
for i in range(num):
    x = int(input())
    if i == 0:
        mx = x
        mi = x
    if x > mx:
        mx = x
    if x < mi:
        mi = x
    print('Максимум - %i, минимум - %i' % (mx, mi))
````
